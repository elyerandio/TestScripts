```python
left_frame = pd.DataFrame({'key' : range(5), 'left_value' : ['a', 'b', 'c', 'd', 'e']})
right_frame = pd.DataFrame({'key' : range(2,7), 'right_value' : ['f', 'g', 'h', 'i', 'j']})
print left_frame
print
print right_frame
```

```python
# inner join (default)
print pd.merge(left_frame, right_frame, on='key', how='inner')
```

```python
# left outer join
print pd.merge(left_frame, right_frame, on='key', how='left')
```

```python
# right outer join
print pd.merge(left_frame, right_frame, on='key', how='right')
```

```python
# full outer join
print pd.merge(left_frame, right_frame, on='key', how='outer')
```

## Combining

## Grouping

```python
headers= ['name', 'title', 'department', 'salary']
chicago = pd.read_csv('city-of-chicago-salaries.csv', names=headers, header=False,
                      converters={'salary': lambda x: float(x.replace('$', ''))})
print chicago.head()
```

```python
by_dept = chicago.groupby('department')
print by_dept
```

```python
# calling count returns the total number of NOT NULL values within each column.
print by_dept.count()
print
print by_dept.size()                     # total records for each department
```

Summation can be done via sum(), averaging by mean(), etc.

```python
print by_dept.sum().head()              # total salaries of each department
print
print by_dept.mean().head()             # average salary of each department
print
print by_dept.median().head()
```

```python
def ranker(df):
    """Assigns a rank to each employee based on salary. with 1 being the highest paid.
    Assumes data is DEC sorted."""
    df['dept_rank'] = np.arange(len(df)) + 1
    return df
```

```python
sort_salary = chicago.sort('salary', ascending=False)
print 'all sorted by salary:\n',sort_salary.head(20)
sort_salary = sort_salary.groupby('department').apply(ranker)
print 'type(sort_salary) :', type(sort_salary)
print 'all sorted by salary with ranking:\n'
sort_salary.head(50)
```

```python
# to print all the highest paid employee per department
sort_salary[sort_salary.dept_rank == 1]
```

```python
# to see employee ranks within their department based on salary
print sort_salary[sort_salary.department == 'LAW'][:25]
```
