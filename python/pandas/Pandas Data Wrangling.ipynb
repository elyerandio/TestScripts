### Remove Values From A Group With Less Than X Rows

```python
from datetime import datetime

def parse(yr, mon):
    date = datetime(year = int(yr), month = int(mon), day = 2)
    return date

url = 'http://www.cpc.ncep.noaa.gov/data/indices/sstoi.indices'
data = pd.read_table(url, sep=' ', header=0, skiprows=0, parse_dates=[['YR', 'MON']], skipinitialspace=True, index_col=0, date_parser=parse)
print data
print data.index
grouped = data.groupby(lambda x : x.year)
```

### Merging on Index

```python
left1 = DataFrame({'key':['a', 'b', 'a', 'a', 'b','c'],
                   'value':range(6) })
right1 = DataFrame({ 'group_val':[3.5, 7]}, index=['a', 'b'])
print 'left1:', '\n', left1, '\n'
print 'right1:', '\n', right1
pd.merge(left1, right1,left_on='key', right_index=True)
```

```python
pd.merge(left1, right1, left_on='key', right_index=True, how='outer')
```

```python
lefth = DataFrame({ 'key1' : ['Ohio', 'Ohio', 'Ohio', 'Nevada', 'Nevada'],
                    'key2' : [2000, 2001, 2002, 2001, 2002],
                    'data' : np.arange(5.) })
righth = DataFrame(np.arange(12).reshape((6,2)),
                   index = [['Nevada', 'Nevada', 'Ohio', 'Ohio', 'Ohio', 'Ohio'],
                            [2001, 2000, 2000, 2000, 2001, 2002]],
                   columns = ['event1', 'event2'])
print 'lefth:\n', lefth,'\n'
print 'righth:\n', righth, '\n'
pd.merge(lefth, righth, left_on=['key1', 'key2'], right_index=True)
```

```python
pd.merge(lefth, righth, left_on=['key1','key2'], right_index=True, how='outer')
```

```python
left2 = DataFrame([ [1., 2.], [3., 4.], [5., 6.]], index=['a', 'c', 'e'],
                  columns=['Ohio', 'Nevada'])
right2 = DataFrame([[7., 8.], [9., 10.], [11., 12.], [13., 14.]],
                   index=['b', 'c', 'd', 'e'],
                   columns=['Missouri', 'Alabama'])
print 'left2:\n', left2, '\n'
print 'right2:\n', right2
pd.merge(left2, right2, left_index=True, right_index=True)
```

```python
pd.merge(left2, right2, left_index=True, right_index=True, how='outer')
```

```python
left2.join(right2, how='outer')
```

```python
left1.join(right1, on='key')
```

```python
another = DataFrame([[7., 8.,], [9.,10.], [11.,12.], [16.,17.]],
                    index=['a','c','e','f'], columns=['New York', 'Oregon'])
print 'left2:', left2, '\n'
print 'right2:', right2, '\n'
print 'another:', another
print left2.join([right2, another])
pd.merge(left2, right2, left_index=True, right_index=True, how='left')
```

### Concatenating Along An Axis

```python
arr = np.arange(12).reshape((3,4))
print 'arr:\n',arr

print '\nconcatenation:\n',np.concatenate([arr, arr])
print '\nconcatenation axis=1:\n', np.concatenate([arr, arr], axis=1)
```

```python
s1 = Series([0,1], index=['a','b'])
s2 = Series([2, 3, 4], index=['c', 'd', 'e'])
s3 = Series([5, 6], index=['f','g'])
print 'concat:\n',pd.concat([s1, s2, s3])
print 'concat axis=1:\n', pd.concat([s1, s2, s3], axis=1)
```

```python
s4 = pd.concat([s1 * 5, s3])
print 's4:\n',s4
print '\nconcat axis=1:\n',pd.concat([s1, s4], axis=1)
print '\nconcat axis=0:\n',pd.concat([s1, s4])
print '\nconcat axis=1 join=inner:\n', pd.concat([s1, s4], axis=1, join='inner')
```
