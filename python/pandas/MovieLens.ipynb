```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
# pass in column names for each CSV
u_cols = ['user_id', 'age', 'sex', 'occupation', 'zip_code']
users = pd.read_csv('ml-100k/u.user', sep='|', names=u_cols)

r_cols = ['user_id', 'movie_id', 'rating', 'unix_timestamp']
ratings = pd.read_csv('ml-100k/u.data',  sep = '\t', names=r_cols)

# the movie file contains columns indicating the movie's genres
# let's only load the first five columns of the file with usecols
m_cols = ['movie_id', 'title', 'release_date', 'video_release_date', 'imdb_url']
movies = pd.read_csv('ml-100k/u.item', delimiter='|', names=m_cols, usecols=range(5))
```

```python
users.tail()
```

```python
ratings.head()
```

```python
movies.head()
```

```python
# create one merged DataFrame
movie_ratings = pd.merge(movies, ratings)
lens = pd.merge(movie_ratings, users)

print lens.head(25)
```

```python
lens.dtypes
```

```python
# 25 most rated movies
print lens.groupby('title').size().order(ascending=False)[:25]
```

```python
# alternatively, pandas has a nifty value_counts method
lens.title.value_counts()[:25]
```

```python
# which movie are most highly rated
movie_stats = lens.groupby('title').agg({'rating' : [np.size, np.mean]})
print movie_stats.head()
```

```python
movie_stats.sort_index(by=[('rating','mean')], ascending=False).head()
```

```python
movie_stats.sort([('rating','mean')], ascending=False).head()
```

```python
# let's look only at movies that have been rated at least 100 times
movie_stats[movie_stats[('rating','size')] >= 100].sort_index(by=[('rating','mean')],ascending=False).head(15)
```

```python
# 50 most rated movie
most_50 = lens.groupby('title').size().order(ascending=False)[:50]
print most_50
```
